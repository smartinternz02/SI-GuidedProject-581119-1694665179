```python
import seaborn as sns
mpg = sns.load_dataset('mpg')
mpg.head()
```

# line plot 

```python
import matplotlib.pyplot as plt
model_years = mpg['model_year']
mpg_values = mpg['mpg']

plt.figure(figsize=(10, 6)) 
plt.plot(model_years, mpg_values, marker='o', linestyle='-')

plt.xlabel('Model Year')
plt.ylabel('MPG')
plt.title('MPG Over the Years')


plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(mpg['mpg'], bins=20, edgecolor='k')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('Distribution of MPG Values')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Scatter Plot of MPG vs. Horsepower')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='origin', y='mpg', data=mpg)
plt.xlabel('Origin')
plt.ylabel('MPG')
plt.title('Box Plot of MPG by Origin')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='origin', y='mpg', data=mpg, ci=None)
plt.xlabel('Origin')
plt.ylabel('Average MPG')
plt.title('Average MPG by Origin')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
origin_counts = mpg['origin'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(origin_counts, labels=origin_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal') 
plt.title('Distribution of Car Origins')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = mpg.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for Auto MPG Dataset')
plt.show()
```
